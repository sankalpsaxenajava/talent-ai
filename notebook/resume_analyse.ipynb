````python
def get_gpt_response(jd_text,resume_text,factors1,factors2):
    system_role = get_system_role(jd_text,resume_text,factors1,factors2)
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "system",
                "content": system_role,
            }
        ],
        model="gpt-4o",
    )
    # print(chat_completion.model)
    return chat_completion.choices[0].message.content

def get_system_role(jd_text,resume_text,factors1,factors2):
    system_role = f"""You are a recruiter whose job is to analyse the resume and assess it for a given Job description. Below are all the factors and maximum score possible for each factor on which you will assess a candidature.
    Always consider the following factors:
    {factors1}
    Consider the following factors only if they are mentioned in the job description. Do not consider them in the calculation if they are not mentioned in the job description.
    {factors2}
    Give me rating for each factor on a scale of 1-10 and also provide a brief explanation for each rating. 
    Calculate the final score out of 100 for the candidature. 
    Job Description Text:
    ```{jd_text}```
    Resume Text:
    ```{resume_text}```
    """    
    # print(system_role)
    return system_role
````

````python
def get_file_text(file_path):
    documents = parser.load_data(file_path)
    text = ""
    for doc in documents:
        text = text + doc.text
    return text
````
