```python
from pydantic import BaseModel, Field
from typing import List, Optional

class Address(BaseModel):
    city: str = ""
    state: str = ""
    country: str = ""
    full_address: str = ""

class Skill(BaseModel):
    skillName: str = ""

class Experience(BaseModel):
    title: str = ""
    company: str = ""
    employment_type: str = ""
    start: str = ""
    end: str = ""
    yearsOfExperience: float = 0.0
    description: str = ""
    industry: str = ""
    location: str = ""
    teamLeadingExperience: bool = False
    skills: List[Skill] = []

class Language(BaseModel):
    name: str = ""
    fluency: str = ""

class Education(BaseModel):
    institution: str = ""
    start_date: str = ""
    end_date: str = ""
    degree_level: str = ""
    degree: str = ""
    field_of_study: str = ""
    specialization: str = ""

class Project(BaseModel):
    title: str = ""
    description: str = ""
    start_date: str = ""
    end_date: str = ""
    techstack: str = ""
    github_link: Optional[str] = None

class CertificationOrTraining(BaseModel):
    name: str = ""
    issuer: str = ""
    issued_date: str = ""
    skills: List[str] = []

class Volunteer(BaseModel):
    organization: str = ""
    position: str = ""
    start: str = ""
    end: str = ""

class Publication(BaseModel):
    title: str = ""
    publisher: str = ""
    release_date: str = ""
    description: str = ""

class Award(BaseModel):
    title: str = ""
    date: str = ""
    awarder: str = ""

class Resume(BaseModel):
    summary: str = ""
    basics: Address = Field(default_factory=Address)
    totalExperienceInYears: float = 0.0
    experience: List[Experience] = []
    softwareToolsOrProgrammingLanguages: str = ""
    languages: List[Language] = []
    education: List[Education] = []
    projects: List[Project] = []
    certificationsAndTrainings: List[CertificationOrTraining] = []
    volunteer: List[Volunteer] = []
    publications: List[Publication] = []
    awards: List[Award] = []
    interests: List[str] = []
    PromotionsOrRoleChangesInOrganization: int = 0
    awardsCount: int = 0
    certificationsOrEducationInNewDomain: bool = False
    significantSkillChange: bool = False
    papersPublishedCount: int = 0
    patentsFiledCount: int = 0
    conferencePresentationsCount: int = 0
    openSourceContributionsCount: int = 0
```
